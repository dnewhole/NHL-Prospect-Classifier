```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_predict = pd.read_csv('../predictions/long/all_train.csv')
valid_predict = pd.read_csv('../predictions/long/all_valid.csv')
```

```python
def adj_proba(row):
    if row.prediction == True:
        return row.confidence
    else:
        return 1 - row.confidence
```

```python
train_predict['adj_conf'] = train_predict.apply(lambda row:adj_proba(row),axis=1)
valid_predict['adj_conf'] = valid_predict.apply(lambda row:adj_proba(row),axis=1)
```

```python
fig = plt.figure(figsize=(18,10))
ax = fig.gca()
plt.scatter(x=train_predict['adj_conf'],y=train_predict['GP'])
plt.xticks(np.arange(0,1.1,step=0.1))
plt.xlabel('Model Confidence')
plt.ylabel('Games Played')
plt.grid()
plt.show()
```

```python
train_predict['n_words'] = train_predict['report'].str.split().apply(len)
```

```python
fig = plt.figure(figsize=(18,10))
ax = fig.gca()
plt.scatter(x=train_predict['adj_conf'],y=train_predict['n_words'])
plt.xticks(np.arange(0,1.1,step=0.1))
plt.xlabel('Model Confidence')
plt.ylabel('Word Count')
plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(18,10))
ax = fig.gca()
plt.scatter(x=train_predict['n_words'],y=train_predict['GP'])
plt.xlabel('Word Count')
plt.ylabel('Games Played')
plt.grid()
plt.show()
```
