```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_predict = pd.read_csv('../predictions/all_train.csv')
```

```python
def adj_proba(row):
    if row.prediction == True:
        return row.probability
    else:
        return 1 - row.probability
```

```python
#train_predict.drop(train_predict[train_predict.GP == 0].index,inplace=True)
train_predict['adj_proba'] = train_predict.apply(lambda row:adj_proba(row),axis=1)
```

```python
fig = plt.figure(figsize=(18,10))
ax = fig.gca()
plt.scatter(x=train_predict['adj_proba'],y=train_predict['GP'])
plt.xlabel('Prediction Probability')
plt.ylabel('Games Played')
plt.grid()
plt.show()
```
